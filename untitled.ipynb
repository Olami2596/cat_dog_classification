```python
#creating a directory for resized images
os.mkdir('image resized')
```

```python
original_folder = 'train/'
resized_folder = 'image resized/'

# List all files in the original folder
file_names = os.listdir(original_folder)

# Separate cat and dog images
cat_files = [f for f in file_names if f.startswith('cat')]
dog_files = [f for f in file_names if f.startswith('dog')]

# Select 1500 from each
cat_files = cat_files[:1500]
dog_files = dog_files[:1500]

selected_files = cat_files + dog_files

# Resize and save to new folder
for filename in selected_files:
    img_path = os.path.join(original_folder, filename)

    with Image.open(img_path) as img:
        img = img.resize((224, 224)).convert('RGB')
        new_img_path = os.path.join(resized_folder, filename)
        img.save(new_img_path)
```

```python
# creaing a for loop to assign labels
filenames = os.listdir('image resized/')


labels = []

for i in range(3000):

  file_name = filenames[i]
  label = file_name[0:3]

  if label == 'dog':
    labels.append(1)

  else:
    labels.append(0)
```

```python
import cv2
import glob
```

```python
image_directory = 'image resized/'
image_extension = ['png', 'jpg']

files = []

[files.extend(glob.glob(image_directory + '*.' + e)) for e in image_extension]

dog_cat_images = np.asarray([cv2.imread(file) for file in files])
```

```python
X = dog_cat_images
Y = np.asarray(labels)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# scaling the data
X_train_scaled = X_train/255

X_test_scaled = X_test/255
```

```python
import tensorflow as tf
import tensorflow_hub as hub
```

```python
mobilenet_model = 'https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'

pretrained_model = hub.KerasLayer(mobilenet_model, input_shape=(224,224,3), trainable=False)
```

```python
from tf_keras.src import models, layers  

num_of_classes = 2

model = models.Sequential()
model.add(pretrained_model)  
model.add(layers.Dense(num_of_classes, activation='softmax'))

model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics = ['acc']
)
```

```python
model.fit(X_train_scaled, Y_train, epochs=5)
```

```python
score, acc = model.evaluate(X_test_scaled, Y_test)
print('Test Loss =', score)
print('Test Accuracy =', acc)
```

```python
efficient_model_url = 'https://www.kaggle.com/models/tensorflow/efficientnet/TensorFlow2/b0-feature-vector/1'

efficient_pretrained_model = hub.KerasLayer(mobilenet_model, input_shape=(224,224,3), trainable=False)

from tf_keras.src import models, layers  

num_of_classes = 2

efficient_model = models.Sequential()
efficient_model.add(efficient_pretrained_model)  
efficient_model.add(layers.Dense(num_of_classes, activation='softmax'))

efficient_model.summary()
```

```python
efficient_model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics = ['acc']
)
```

```python
efficient_model.fit(X_train_scaled, Y_train, epochs=5)
```

```python
score, acc = efficient_model.evaluate(X_test_scaled, Y_test)
print('Efficient Model Test Loss =', score)
print('Efficient Model Test Accuracy =', acc)
```
